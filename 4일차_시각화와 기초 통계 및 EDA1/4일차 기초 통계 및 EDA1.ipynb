## 구간 추정

* 구간 추정을 하기 위해서는 모수를 포함하는 구간의 확률을 지정해야 함
    * 일반적으로 95% or 99%, 해당 값은 임의의 값이며 신뢰계수라고 함
    * 신뢰계수에 대한 구간 = 신뢰하한 $\leq$ 모수 $\leq$ 신뢰상한
        * 하한과 상한은 표본 통계량을 통해서만 도출, 이는 추정하고자 하는 모수가 알맞은 추정구간 내에 존재한다고 믿을 수 있다는 정도를 의미
    * 구간 추정에서는 신뢰수준, 신뢰구간을 함께 제시

* 신뢰계수가 높고 범위가 적은 것이 이상적
    * 계수가 높아질수록 범위는 커짐
    * 반대로 계수가 낮아질수록 범위는 작아짐
    * 이율배반적 관계
    * 하지만 데이터가 늘어날수록 범위는 작아짐

## 확률

* 표본공간(S): 얻을 수 있는 가능한 모든 결과들의 전체 집합
* 사건: 어떤 조건에 만족하는 결과들의 집합 $\rightarrow$ 표본공간의 부분집합
* $확률(A) = \frac{원하는 결과의 경우의 수}{가능한 결과의 모든 경우의 수} = \frac{사건 A의 원소의 개수}{표본공간 S의 원소의 개수}$
* 2개의 동전이 존재하고, 이를 던졌을 때 모두 앞면인 확률은?
    * S = {HH, HT, TH, TT}
    * A = {HH}

* 확률 변수
    * 무작위 실험을 했을 때, 특정 활률로 발생하는 각각의 결과를 수치적 값으로 표현되는 변수
    * 확률 변수(X) = 앞면(1), 뒷면(0)

````python
import pandas as pd

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
'''
fig, axes = plt.subplots(1, 2)
plt.subplots_adjust(wspace=0.35)
axes[0].set_title("Train")
axes[1].set_title("Test")

plt.xticks(np.arange(0, 10))
sns.countplot(x=train['label'], ax=axes[0])
sns.countplot(x=test['label'], ax=axes[1]);
'''
````

````python
train_x = np.array(train.iloc[:,:784])
train_x = train_x / 255
train_x = train_x.reshape(-1, 28, 28, 1)
train_y = np.array(train['label'])

test_x = np.array(test.iloc[:,:784])
test_x = test_x / 255
test_x = test_x.reshape(-1, 28, 28, 1)
test_y = np.array(test['label'])

bin_train_y = (train_y == 0)
bin_test_y = (test_y == 0)

tt = [1 if i else 0 for i in bin_train_y]

plt.xticks(np.arange(0, 10))
plt.hist(tt)
plt.show()
````

````python
np.sum(tt)
````

````python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(train_x, bin_train_y, test_size=0.2, random_state=34)
````

````python
#pip install tensorflow
````

````python
from tensorflow.keras.layers import *
from tensorflow.keras import Sequential
from tensorflow.keras import optimizers

model=Sequential()
model.add(Conv2D(filters=64, kernel_size = (3,3), activation="relu", input_shape=(28,28,1)))
model.add(Conv2D(filters=64, kernel_size = (3,3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization())

model.add(Conv2D(filters=128, kernel_size = (3,3), activation="relu"))
model.add(Conv2D(filters=128, kernel_size = (3,3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization())    

model.add(Conv2D(filters=256, kernel_size = (3,3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(BatchNormalization())
    
model.add(Flatten())
model.add(Dense(512,activation="relu"))

model.add(Dense(units=1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer= optimizers.Adam(learning_rate=0.001), metrics=['accuracy'])

hist = model.fit(x_train, y_train, epochs=5, batch_size=128, validation_data=(x_valid, y_valid))
````

````python
prediction = model.predict(test_x)
prediction = np.round(prediction, 0).reshape(-1)
table = (bin_test_y == prediction)
len(table[table == True]) / len(bin_test_y)
````

````python
c_matrix = np.zeros([2, 2]).astype(np.int64)

for i in range(bin_test_y.shape[0]):
    idx_target = np.abs(bin_test_y[i].astype(int) - 1)
    idx_output = np.abs(prediction[i].astype(int) - 1)

    c_matrix[idx_target, idx_output] += 1

c_matrix_frame = pd.DataFrame(c_matrix, index=['Actual(P)', 'Actual(N)'], columns=['Predicted(P)', 'Predicted(N)'])
````

````python
import seaborn as sns

sns.heatmap(c_matrix_frame, annot = True, fmt = 'd',cmap = 'Blues')
plt.title('Shuffle data')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.xticks([0.5, 1.5],['P', 'N'])
plt.yticks([0.5, 1.5],['P', 'N'])
plt.show()
````

## 네이만-피어슨 추론 방식

* 표본으로 모집단을 추정하는 일반적인 방법
    * 가설 수립
        * 현재 알지 못하는 모집단에 대해 가설 설정
        * 가설을 증명하기 위한 표본 수집 및 분석
    * 표본이 유의할 확률 계산
        * 표본이 모집단을 대표할 수 있는지 확인
        * 분석한 표본이 관찰되는 이유는 그것이 모집단 내에서 일반적이라는 점에서 시작
            * 모집단 내에서는 흔한 사건
            * 이를 유의성이라고 함
    * 가설 기각/채택

## 귀무가설과 대립가설

* 귀무가설($H_0$)
    * 알고자 하는 모집단의 특성에 대한 잠정적인 주장
    * 보편적으로 알려진 기존의 사실
    * A와 B는 관계 혹은 차이가 없다. 혹은 서로 영향을 주지 않는다.
* 대립가설($H_1$)
    * 귀무가설을 기각하고 그것에 대한 대안으로 채택하게 되는 가설
    * 나의 실험이 의미 있는 새로운 사실을 밝혀냈다는 주장
    * A와 B는 관계 혹은 차이가 있다. 혹은 서로 영향을 준다.
    

* 귀무가설을 검정의 대상으로 삼는 이유

## 유의 확률

* 유의 확률($p$)
    * 귀무가설을 지지하는 힘
    * p-value는 귀무가설이 맞다는 전제 하에, 표본에서 실제로 관측된 통계치와 같거나 더 극단적 통계치가 관측될 확률을 나타냄
    * 관찰된 결과가 귀무가설과 양립하는 정도를 0 ~ 1사이의 값으로 표현
    * p-value가 작을수록 그 정도가 약하다고 보며, 유의수준 보다 작을 경우 귀무가설을 기각
        * 유의수준은 일반적으로 0.05
    * 대립가설을 지지하는 검정통계치를 모을 확률로 귀무가설을 지지하는 정도로 해석
    * 만약 유의 확률이 0.01이라면 귀무가설을 기각했을 때 이 기각 결정이 잘못된 확률은 1%라는 것이되며, 이와 같이 이번 기각 졀정이 잘못될 확률이 너무 작기 때문에 과감히 기각된다고 해석 가능

### T-test

* 두 집단간 평균 차이가 유의미한지 검증하는 방법

* T-test 수행을 위한 전제 조건
    1. 표본이 독립적인가?
    2. 수집된 데이터가 정규분포를 따르는가?
    3. 비교 집단이 2개인가?

* Paired T-test 수행을 위한 전제 조건
    1. 표본이 비독립적인가?
    2. 수집된 데이터가 정규분포를 따르는가?
    3. 비교 집단이 2개인가?

    * 한 햄버거 가게의 패티의 중량은 최소 200g이라고 주장 = 귀무가설
        * 샘플 $n$ = 20
        * 평균($\bar{x}$) = 196
        * 표준편차($s$) = 5.3
        * 유의수준(0.05)

<img src="https://cdn.graphpad.com/assets/0.74.0/images/srcset/t-test-ratio-w622-300.png" width="150">

````python
import scipy.stats as st

n, mu, x_bar, s, alpha = 20, 200, 196, 5.3, 0.05

t_statistic = (x_bar - mu) / (s / np.sqrt(n))
t_alpha = st.t.ppf(alpha, n-1)
p_val = st.t.sf(np.abs(t_statistic), n-1)
print('검정 통계량:', round(t_statistic, 3))
print('임계값:', round(t_alpha, 3))
print('유의확률:', round(p_val, 3))
print('귀무가설 기각') if p_val < alpha else print('대립가설 기각')
````
