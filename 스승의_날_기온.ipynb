```python
# 매년 스승의 날인 5월 15일을 기준으로 최고 기온 데이터에 대한 그래프 그리기
filename = '/content/drive/MyDrive/Colab Notebooks/busan.csv'

import csv
f = open(filename, 'r', encoding='cp949')
data = csv.reader(f)
next(data)
result = [] # 최고 기온 데이터를 저장할 리스트 생성
for row in data:
  if row[2] != '':
    if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '15':
      result.append(float(row[2]))

print(len(result))

f.close()

plt.plot(result, 'r')
plt.show()
```

```python
# 매년 스승의 날인 5월 15일을 기준으로 최저 기온 데이터에 대한 그래프 그리기
filename = '/content/drive/MyDrive/Colab Notebooks/busan.csv'

import csv
f = open(filename, 'r', encoding='cp949')
data = csv.reader(f)
next(data)
result = []
lresult = []
for row in data:
  if row[2] != '' and row[4] !='':
    if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '15':
      result.append(float(row[2]))
      lresult.append(float(row[4]))
print(len(result))

f.close()
plt.title("스승의 날")
plt.plot(result, 'red', label='최고기온')
plt.plot(lresult, 'blue',label='최저기온')
plt.legend()
plt.show()
```
